```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Import data to notebook as "data":
data = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
print(data.head())
print(data.tail())
```

```python
# Max Price:
print('Brand with Max Price:')
data[data['Price'] == data['Price'].max()]
```

```python
# Missing Values:
# pd.isnull(data)
pd.isnull(data).sum()
```

```python
# Replacing missing values:
min_mean = data['Min.Price'].mean()
max_mean = data['Max.Price'].mean()
values = {'Min.Price': min_mean, 'Max.Price': max_mean}
data.fillna(values, inplace = True)
data
```

```python
# Sorting cars based on 'Price':
data.sort_values(by = 'Price', inplace = True, ascending = True)
data.reset_index(drop = True, inplace = True)     # Reindexing
data
```
